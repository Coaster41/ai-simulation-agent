```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('results/results_100_10000.csv', header=None)
results = df.to_numpy()
X=results[:100,:-1]
y=(results[:100,-1]-0.5)*10
```

```python
x_unique = np.unique(X)
num_units = (len(x_unique)-1)//2
costs = np.zeros(num_units+1)
print(costs)
```

```python
(keys, values) = np.unique(X[0], return_counts=True)
neg_index = np.nonzero(keys<0)
neg_cost = np.sum(costs[np.abs(keys[neg_index]).astype(int)]*values[neg_index])
pos_index = np.nonzero(keys>0)
pos_cost = np.sum(costs[np.abs(keys[pos_index]).astype(int)]*values[pos_index])
delta_cost = pos_cost-neg_cost
```

```python
def MSE(X, y, costs):
    total_error = 0
    for row in range(X.shape[0]):
        (keys, values) = np.unique(X[row], return_counts=True)
        neg_index = np.nonzero(keys<0)
        neg_cost = np.sum(costs[np.abs(keys[neg_index]).astype(int)]*values[neg_index])
        pos_index = np.nonzero(keys>0)
        pos_cost = np.sum(costs[np.abs(keys[pos_index]).astype(int)]*values[pos_index])
        delta_cost = pos_cost-neg_cost
        total_error += (y[row]-delta_cost)**2
        # if row%1000 == 0:
        #     print(row)
    return total_error/X.shape[0]
```

```python
print(MSE(X, y, costs))
```
